2. Модифицируйте функцию разбора JSON таким образом, чтобы в каждый словарь c информацией об отзыве добавить ключ `rating`. Значение получите на основе названия файла (см. аргумент `include_path`), использовав для этого регулярное выражение.

```python
import dask.bag as db
import json
import re

# Функция для разбора JSON и добавления ключа rating на основе названия файла
def parse_json(json_str):
    rating = int(re.search(r'reviews_(\d).json', json_str[1]).group(1))
    review = json.loads(json_str[0])
    review['rating'] = rating
    return review

# Считываем файлы в виде Dask Bag и применяем функцию parse_json
reviews_bag = db.read_text('/content/drive/MyDrive/reviews_full/reviews_full/reviews_*.json', include_path=True).map(json.dumps).map(json.loads).map(parse_json)

# Выводим первые 5 элементов Dask Bag
print(reviews_bag.take(5))
```

3. Посчитайте количество отзывов в исходном датасете.

```python
reviews_bag.count().compute()
```

4. Отфильтруйте `bag`, сохранив только отзывы, оставленные в 2014 и 2015 годах.

```python
reviews_filtered = reviews_bag.filter(lambda x: x['date'].startswith('2014') or x['date'].startswith('2015'))
reviews_filtered.take(5)
```

5. Выполните препроцессинг отзывов:
    * привести строки к нижнему регистру
    * обрезать пробельные символы в начале и конце строки
    * удалите все символы, кроме английских букв и пробелов
    
Примените препроцессинг ко всем записям из `bag`, полученного в задании 4.

```python
def preprocess_review(obj):
    obj['review'] = obj['review'].lower()
    obj['review'] = obj['review'].strip()
    obj['review'] = re.sub(r'[^\w\s]', '', obj['review'])
    return obj

preprocessed_bag = reviews_filtered.map(preprocess_review)
preprocessed_bag.take(5)
```

6. Посчитайте количество отзывов в датасете, полученном в результате решения задачи 5. В случае ошибок прокомментируйте результат и исправьте функцию препроцессинга.

```python
preprocessed_bag.count().compute()
```

7. Посчитайте, как часто в наборе, полученном в задании 5, встречается та или иная оценка

```python
ratings_freq = preprocessed_bag.map(lambda x: x['rating']).frequencies().compute()
print(ratings_freq)
```

8. Найдите среднее значение `rating` в выборке

```python
ratings_bag = preprocessed_bag.pluck('rating')
mean_rating = ratings_bag.mean().compute()
print(mean_rating)
```

9. Используя метод `foldby`, подсчитать максимальную длину отзывов в зависимости от оценки `rating` в наборе, полученном в задании 5.

```python
max_len_by_rating = preprocessed_bag.foldby('rating', lambda x, y: max(x, len(y['review'])), 0).compute()
print(max_len_by_rating)
```
