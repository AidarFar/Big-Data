2) Датасет: - Необходимо найти все целые числа от 1 до 400 000, для которых sin от этого числа будет иметь не менее 9 одинаковых цифр (например, sin(139198)=0.30763333543133703; sin(139198)=0.30763333543133703 т.е. имеет 9 троек). Ответ представляет собой список пар (целое число, значение синуса для него), для которых значение синуса удовлетворяет условию. Решение этой задачи распараллелить, используя multiprocessing Pool. Сравнить продолжительность последовательного и параллельного решения задачи.

````python
import math
from multiprocessing import Pool

def check_sine(n):
    sin_value = math.sin(n)
    sin_str = str(sin_value)
    digit_counts = [sin_str.count(digit) for digit in sin_str if digit.isdigit()]
    if max(digit_counts) >= 9:
        return (n, sin_value)
    return None

def find_numbers_parallel():
    numbers = range(1, 400001)
    with Pool() as pool:
        results = pool.map(check_sine, numbers)
    return [result for result in results if result is not None]

def find_numbers_sequential():
    numbers = range(1, 400001)
    return [check_sine(n) for n in numbers if check_sine(n) is not None]
````

````python
%%time
sequential_results = find_numbers_sequential()
print("Sequential results:", sequential_results)
````

````python
%%time
parallel_results = find_numbers_parallel()
print("Parallel results:", parallel_results)
````

3) Датасет: Chinook_Sqlite.sqlite С помощью кода на Python с использованием sqlite3 и SQL решить задачу. Реализовать функции на Python: Которая возвращает всех имеющихся артистов. Которая по имени артиста возвращает все его альбомы с количеством треков в них.

````python
import sqlite3

conn = sqlite3.connect('/content/drive/MyDrive/Chinook_Sqlite.sqlite')
````

````python
def get_all_artists():
    query = "SELECT Name FROM Artist"
    cursor = conn.execute(query)
    artists = [row[0] for row in cursor.fetchall()]
    return artists
````

````python
def get_albums_by_artist(artist_name):
    query = """
    SELECT Album.Title, COUNT(Track.TrackId) AS TrackCount
    FROM Artist
    JOIN Album ON Artist.ArtistId = Album.ArtistId
    JOIN Track ON Album.AlbumId = Track.AlbumId
    WHERE Artist.Name = ?
    GROUP BY Album.AlbumId
    """
    cursor = conn.execute(query, (artist_name,))
    albums = cursor.fetchall()
    return albums
````
