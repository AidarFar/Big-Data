4) Построить "one-hot encoding" для одномерного массива numpy из целых неотрицательных чисел (длина массива и максимальное значение в массиве заранее неизвестно). Протестировать свое решение на случайно сгенерированном одномерном массиве, соответствующем условию.Решить задачу средствами numpy и/или pandas. Не использовать циклы и конструкции стандартного Python там, где можно использовать возможности данных библиотек.

```python
import numpy as np

def one_hot_encoding(arr):
    unique_vals, encoded_arr = np.unique(arr, return_inverse=True)
    encoded = np.eye(len(unique_vals))[encoded_arr]
    return encoded

arr = np.random.randint(0, 1000000, size=10)
encoded_arr = one_hot_encoding(arr)

print("Исходный массив:")
print(arr)
print("\nOne-hot encoding:")
print(encoded_arr)

```
