1. Напишите 2 функции, имитирующие CPU-bound задачу и IO-bound задачу:

`cpu_task()`: генерирует 100 тыс. случайных чисел и возвращает их сумму (без использования `numpy`)

`io_task()`: "спит" 0.1 сек, затем генерирует случайное число и возвращает его

Замерьте время выполнения 100 последовательных вызовов каждой из этих функций. Распараллелив вычисления при помощи `dask.delayed`, сократите время выполнения. Исследуйте, как зависит время вычислений от выбранного планировщика `scheduler`.

```python
import random
import time

def cpu_task():
    return sum(random.randint(0, 100) for _ in range(100000))


def io_task():
    time.sleep(0.1)
    return random.randint(0, 100)
```

```python
%%time
for _ in range(100):
  cpu_task()
  io_task()
```

```python
%%time
import dask
dask.config.set(scheduler='threads')

results=[]
for _ in range(100):
    results.append(dask.delayed(cpu_task)())
    results.append(dask.delayed(io_task)())

dask.compute(*results)
```

```python
%%time
dask.config.set(scheduler='processes')

results=[]
for _ in range(100):
    results.append(dask.delayed(cpu_task)())
    results.append(dask.delayed(io_task)())

dask.compute(*results)
```

```python
%%time
dask.config.set(scheduler='synchronous')

results=[]
for _ in range(100):
    results.append(dask.delayed(cpu_task)())
    results.append(dask.delayed(io_task)())

dask.compute(*results)
```

```python
%%time
dask.config.set(scheduler='single-threaded')
results=[]
for _ in range(100):
    results.append(dask.delayed(cpu_task)())
    results.append(dask.delayed(io_task)())

dask.compute(*results)
```

Напишите функцию, которая считывает файл формата xml из архива `reviewers_full.zip` и по данным этого файла формирует список словарей, содержащих следующие ключи: `username`, `name`, `sex`, `country`, `mail`, `registered`, `birthdate`, `name_prefix`, `country_code`. Часть из этих значений в исходном файле хранится в виде тэгов, часть - в виде атрибутов тэгов. Для конкретного человека какие-то из этих ключей могут отсутствовать.



```python
import xml.etree.ElementTree as ET

def read_xml_file(file_path):
    tree = ET.parse(file_path)
    root = tree.getroot()

    data = []

    for user in root:
        user_dict = {}

        # Read data from tags
        user_dict['id'] = user.findtext('id')
        user_dict['username'] = user.findtext('username')
        user_dict['name'] = user.findtext('name')
        user_dict['sex'] = user.findtext('sex')
        user_dict['country'] = user.findtext('country')
        user_dict['mail'] = user.findtext('mail')
        user_dict['registered'] = user.findtext('registered')
        user_dict['birthdate'] = user.findtext('birthdate')


        # Read data from attributes
        user_dict['name_prefix'] = user.get('prefix')
        user_dict['country_code'] = user.get('code')

        data.append(user_dict)

    return data
```

3. Задекорируйте функцию из задания 1 при помощи `dask.delayed` и создайте список `reviewers`, состоящий из 5 объектов `delayed` (по одному объекту на файл). Из списка объектов `delayed`, создайте `dask.bag` при помощи метода `db.from_delayed`. Добавьте ключ `birth_year`, в котором хранится год рождения человека. Оставьте в выборке только тех людей, которые __наверняка__ моложе 1980 года. Преобразуйте поле `id` к целому типу.

```python
import xml.etree.ElementTree as ET
import dask.bag as db
@dask.delayed
def read_xml_file(file_path):
    tree = ET.parse(file_path)
    root = tree.getroot()

    data = []

    for user in root:
        user_dict = {}

        # Read data from tags
        user_dict['id'] = user.findtext('id')
        user_dict['username'] = user.findtext('username')
        user_dict['name'] = user.findtext('name')
        user_dict['sex'] = user.findtext('sex')
        user_dict['country'] = user.findtext('country')
        user_dict['mail'] = user.findtext('mail')
        user_dict['registered'] = user.findtext('registered')
        user_dict['birthdate'] = user.findtext('birthdate')


        # Read data from attributes
        user_dict['name_prefix'] = user.get('prefix')
        user_dict['country_code'] = user.get('code')
        birthdate = user_dict['birthdate']
        if birthdate:
            user_dict['birth_year'] = int(birthdate.split('-')[0])
        else:
            user_dict['birth_year'] = None
        data.append(user_dict)
        user_dict['id'] = int(user_dict['id']) if user_dict['id'] else None
    return data
```

```python
reviewers = [dask.delayed(read_xml_file)(file) for file in xml_files]

# reviewers = list(chain.from_iterable(reviewers))
bag_reviewers = db.from_delayed(reviewers)

filtered_reviewers = bag_reviewers.filter(lambda person: person['birth_year'] is not None and person['birth_year'] > 1980)
```
